```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
        
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
plt.plot(x, y)
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight')
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1, np.max(y.cumsum()) + 1)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'r*')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title("Number of Hamburgers")
```
